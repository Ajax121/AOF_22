```python
inp = open('1.txt','r').readlines()
```

```python
import re
lst = []
for num in inp:
    lst.append((re.sub('\n','',num)))
```

```python
lst.append('')
```

```python
lst[-1]
```

```python
lst_of_lsts = []
sublist = []
for item in lst:
    if item != "":
        sublist.append(int(item))
    else:
        lst_of_lsts.append(sum(sublist))
        sublist = []
if sublist != []:
     lst_of_lsts.append(sublist)
```

```python
max(lst_of_lsts)
```

```python
lst_of_lsts.sort()
```

```python
lst_of_lsts
```

```python
sum(lst_of_lsts[::-1][:3])
```
